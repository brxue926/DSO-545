```python
import pandas as pd
```

```python
data = pd.read_csv("trump_approval.csv")
```

```python
data.head()
```

```python
# Question 1
# Update the datat type of the Date variable to a date/time variable.

data["Date"]= pd.to_datetime(data["Date"]) 
data.head()
```

```python
# Question 2
# Create a line chart that shows Presidentâ€™s Trump approval rating.

import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (11,6))
plt.plot(data.Date,data.Approve,
         linestyle = "-",
         color = "green")
plt.plot(data.Date, data.Disapprove,
         linestyle = "-",
         color = "orange")
plt.title("President Trump's Approval Rating",fontsize = 14,fontweight = "bold")
plt.ylabel("President's Rating(%)", fontsize = 10)
plt.show()
```

```python
# Question 6

plt.figure(figsize = (11,6))
plt.plot(data.Date,data.Approve,
         linestyle = "-",
         color = "green")
plt.plot(data.Date, data.Disapprove,
         linestyle = "-",
         color = "orange")

plt.ylim(30,70) 
plt.title("President Trump's Approval Rating",fontsize = 14,fontweight = "bold")
plt.ylabel("President's Rating(%)", fontsize = 10)

plt.axvline(x = '2018-01-01', linestyle = "--", color = "grey")
plt.axhline(y = 50, linestyle = ":", color = "red")
plt.axvspan("2018-12-22","2019-01-25",facecolor='grey', alpha=0.3)
plt.show()
```

```python
# Question 7

plt.figure(figsize = (11,6))
plt.plot(data.Date,data.Approve,
         linestyle = "-",
         color = "green",
         label = "Approval")
plt.plot(data.Date, data.Disapprove,
         linestyle = "-",
         color = "orange",
         label = "Disapproval")

plt.ylim(30,70) 
plt.title("President Trump's Approval Rating",fontsize = 14,fontweight = "bold")
plt.ylabel("President's Rating(%)", fontsize = 10)

plt.axvline(x = '2018-01-01', linestyle = "--", color = "grey")
plt.text('2018-01-01', 65, 'One year \n in office', fontsize= 12)
plt.axhline(y = 50, linestyle = ":", color = "red")
plt.axvspan("2018-12-22","2019-01-25",facecolor='grey', alpha=0.3)
plt.text("2018-12-27",65,"Dec.22 - Jan.25\nGovernment shutdown", fontsize= 12)
plt.legend(loc ='upper left')
plt.text("2019-08-01",57,"57%\nDisapprove", fontsize = 12, color = 'orange')
plt.text("2019-08-01",39,"39%\nApprove", fontsize = 12, color = 'green')
plt.show()
```
